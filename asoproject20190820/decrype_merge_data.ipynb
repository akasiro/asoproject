```python
conn = sqlite3.connect(r'D:\documents\data\aso.niaoge\20190723\aso20190723.db')
cur = conn.cursor()
testconn = sqlite3.connect(r'D:\documents\Buffer\test2.db')
testcur = testconn.cursor()
```

```python
pdmg_var_dict = {
    'datahash': 'varchar(100) primary key',
    'pkgName':'varchar(50)',
    'appHash':'varchar(50)',
    'id':'varchar(50)',
    'type':'varchar(20)',
    'type_code':'varchar(20)',
    'date':'varchar(20)',
    'user_num':'integer',
    'active_num':'integer',
    'cover_num':'integer',
    'mau': 'float',
    'coverage':'float','g_app_m':'float',
    'g_app_v':'float',
    'g_apptype_m':'float',
    'g_apptype_v':'float',
    'g_gtype_m':'float',
    'g_gtype_v':'float',
    'g_game_m':'float',
    'g_game_v':'float',
    'g_rgame_m':'float',
    'g_rgame_v':'float',
    'g_rgtype_m':'float',
    'g_rgtype_v':'float'
}

pdmg_var_define = ','.join(['{} {}'.format(k,v) for k,v in pdmg_var_dict.items()])
create_table_pdmg_cmd = 'create table if not exists game_demand_perf_td_20190820 ({})'.format(pdmg_var_define)
print(create_table_pdmg_cmd)

try:
    cur.execute(create_table_pdmg_cmd)
    conn.commit()
except:
    print('wrong in creating table')

pdmg2columns = pdmg2.columns.tolist()
hash_sha256 = hashlib.sha256()
for r in pdmg2.itertuples(index = False):
    hash_sha256.update(str(tuple(r)).encode())
    datahash = hash_sha256.hexdigest()
    tempcmd = 'insert into game_demand_perf_td_20190820(datahash,{}) values ({}?)'.format(','.join(pdmg2columns),'?,'*(len(pdmg2columns)))
    try:
        cur.execute(tempcmd,(datahash,)+tuple(r))
        conn.commit()
    except:
        print('wrong')
```
