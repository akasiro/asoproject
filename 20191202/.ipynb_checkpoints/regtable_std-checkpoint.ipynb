```python
# std func
def pd_std(df,columns):
    df = df.copy()
    for c in columns:
        df['{}_std'.format(c)] = (df[c]-np.mean(df[c]))/np.std(df[c])
    df_std = df.drop(columns = columns)
    return df_std,df
```

```python
# subset type_time data
# include demand heterogeneity in category level
type_time = baseregtable[['type_code', 'date', 'timestamp',
                          'app_m', 'app_v', 'apptype_m', 'apptype_v',
                          'gtype_m', 'gtype_v','game_m', 'game_v', 
                          'rgame_m', 'rgame_v', 'rgtype_m', 'rgtype_v',
                          'type_ratio_m', 'type_ratio_v']].drop_duplicates()

type_time_std_columns = ['app_m', 'app_v', 'apptype_m', 'apptype_v',
                          'gtype_m', 'gtype_v','game_m', 'game_v', 
                          'rgame_m', 'rgame_v', 'rgtype_m', 'rgtype_v',
                          'type_ratio_m', 'type_ratio_v']
type_time_std = pd_std(type_time,type_time_std_columns)[0]
print(type_time_std.shape,'\n')
type_time_std.head()
```
